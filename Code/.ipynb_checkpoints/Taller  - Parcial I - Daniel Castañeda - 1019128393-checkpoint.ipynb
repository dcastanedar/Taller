#### Punto 3.

* <div align="justify">Dado un autómata finito determinístico $M$, construya un nuevo autómata finito determinístico $M'$ con el mismo alfabeto de entrada y con la propiedad de que para todos los , &mu; &in; &Sigma;&#42;, es aceptado por $M'$ si y  solo si NO es aceptado por $M$ .<br><br>
Se define un autómata $M$ que es de la forma:<br><br>

<center>$M$ = ( Q,  &Sigma;&#42;,  &delta;,  q<sub>0</sub>,  F)<br></center><br>

Y que el lenguaje aceptado por $M$ está definido como:<br><br>
<center>$L(M)$ = (w &in; &Sigma; &#42; | &delta;(q<sub>0</sub>, w) &in; F)<br></center><br>

Se tiene que &Sigma;&#42;, corresponde a todas las posibles palabras formadas con un alfabeto &Sigma;. $L(M)$ es un subconjunto de &Sigma;&#42;, y cuyo complemento serán las palabras que no acepta el autómata $M$.<br><br>

<center><i>&Sigma;&#42; = $L(M)$ U {W &in; &Sigma;&#42; | &delta;(q<sub>0</sub>, w) &notin; F}</i><br></center><br>

Donde &delta; corresponde a la función de transición de $M$. Para una palabra $w$ donde <i>&delta;(q<sub>0</sub>, $w$) &notin; F  </i>, aquel estado que arroja la función de transición, hará parte del conjunto de estados finales para un nuevo automata $M'$, llamado F', definido también como:   <br><br>

<center><i>&delta;(q<sub>0</sub>, $w$) &in; F' </i></center><br>

Con esto, podemos decir que $M'$ acepta todas las palabras que no son aceptadas en $M$, ya que los estados finales de $M'$ corresponden a los estados no finales de $M$.   <br><br>
Para concluir, se define el autómata $M'$ como:<br>
<center>$M'$ = ( Q,  &Sigma;,  &delta;,  q<sub>0</sub>,  F')<br></center><br></div>



#### Punto 4.
* <div align="justify"> <h>Dados dos autómatas finitos determinísticos <i>M</i><sub>1</sub> y <i>M</i><sub>2</sub> con el mismo alfabeto de entrada, construya un tercer autómata finito determinístico (<i>M</i>) con la propiedad &mu; &isin; &Sigma;* que es aceptado por <i>M</i> si y solo si es aceptado por ambos  <i>M</i><sub>1</sub> y <i>M</i><sub>2</sub>.  </h><br>

<br>Dado el problema, se tiene que:<br><br>

<center><i>M</i><sub>1</sub> = { Q<sub>1</sub>,  &Sigma;,  &delta;<sub>1</sub>,  q<sub>1</sub>,  F<sub>1</sub>}<br><br> y <br><br>
<i>M</i><sub>2</sub> = { Q<sub>2</sub>,  &Sigma;,  &delta;<sub>2</sub>,  q<sub>2</sub>,  F<sub>2</sub>}<br></center>

Se define un tercer autómata <i>M</i><sub>3</sub>, que es de la forma:<br><br>

<center><i>M</i><sub>3</sub> = { Q<sub>3</sub>,  &Sigma;,  &delta;<sub>3</sub>,  q<sub>3</sub>, F<sub>3</sub>}<br><br></center>

Se tiene que Q<sub>3</sub> se define como el producto cartesiano entre Q<sub>1</sub> y Q<sub>2</sub>, donde en Q<sub>3</sub> será el conjunto de parjeas ordenadas de la forma <i>(q<sub>1</sub>,q<sub>2</sub>)</i>, para todo <i>q<sub>1</sub> &in; Q<sub>1</sub></i> y para todo <i>q<sub>2</sub> &in; Q<sub>2</sub></i> : <br><br>

<center>Q<sub>3</sub>  =  Q<sub>1</sub>  x  Q<sub>2</sub><br><br></center>

Además se define una función de transición para M<sub>3</sub> basada en las funciones de transición de M<sub>1</sub> y M<sub>2</sub> :<br><br>
<center>&delta;<sub>1</sub>(  i,  a ) = ( <i>q<sub>f1</sub></i> )<br>
&delta;<sub>2</sub>(  j,  a ) = ( <i>q<sub>f2</sub></i> )<br>
&delta;<sub>3</sub>( ( i,  j ),  a ) = ( <i>q<sub>f1</sub></i>,  <i>q<sub>f2</sub></i> )<br><br></center>

Donde  a corresponde a un símbolo perteneciente a &Sigma;  y donde en la función de transición &delta;<sub>3</sub>, cambia el parámetro de estado de partida, por la pareja ordenada que pertenece a Q<sub>3</sub>, y como resultado arroja una pareja ordenada de estados finales para todo <i>q<sub>f1</sub></i> &in; F<sub>1</sub> y  para todo <i>q<sub>f2</sub></i> &in; F<sub>2</sub>.<br><br> Una pareja ordenada que pertenezca a Q<sub>3</sub> puede representarse de la forma:<br><br>
<center><i>q<sub>f3</sub></i>  =  ( <i>q<sub>f1</sub></i>,  <i>q<sub>f2</sub></i> )<br><br></center>

A la vez, dada esta función de transición, se tiene que F<sub>3</sub> se defne como:<br>
<center>F<sub>3</sub>  =  F<sub>1</sub>  x  F<sub>2</sub><br><br></center>
Y así se concluye que M<sub>3</sub> es un autómata cuyo lenguaje es el mismo, tanto de M<sub>1</sub> como de M<sub>2</sub>.</div>

#### Punto 7.
* <div align="justify"> <h>Un sistema de transiciones es una generalización de los autómata finito no
determinísticos con  &epsilon;-transiciones, en la cual se admiten transiciones adicionales con
palabras de longitud mayor o igual a 1. De una definición formal de los sistemas de
transiciones, y pruebe que los lenguajes aceptados por un sistema de transiciones son
aceptados por un DFA.  </h><br>

#### Definición Formal: <br>

Formalmente, un sistema de transiciones es un par ($S$, &rarr;), donde $S$ es un conjunto de estados y &rarr; es un conjunto de transiciones de estado, (conocido como un subconjunto de S x S). El hecho de que exista una transición desde un estado $p$ a un estado $q$, es decir, $(p,q)$ &in; &rarr;, esta escrito como $p$ &rarr; $q$.<br>

Un sistema de transición etiquetado es una tupla de la forma ($S$, &and;, &rarr;), donde $S$ es un conjunto de estados, &and; es un conjunto de etiquetas y &rarr; es un conjunto de transiciones etiquetadas (es decir, un subconjunto de $S$ x &and; x $S$). La tupla ($p$, &alpha;, $q$) &in; &rarr; representa el hecho de que es una transición desde un estado $p$ a un estado $q$ con etiqueta &alpha;.<br>

Las etiquetas pueden representar diferentes cosas, dependiendo en el lenguaje definido, que usualmente representan una entrada esperada, para que la transición sea efectuada.<br>

Si para algun estado $p$ y un &alpha;, existe una única tupla ($p$, &alpha;, $q$) en &rarr;, entonces se dice que &alpha; es determinístico para $p$, lo que significa que los lenguajes aceptados por estos sistemas son aceptados por un autómata finito determinístico.

Dado este autómata, se procede a demostrar la afirmación por contraejemplo.<br>
Se tiene una palaba $w = ac$, donde según la afirmación, el autómata debería aceptar. Al hacer la función de transición de esta palabra se tiene que: <br><br>
<center>($1$) &delta;(1, a) = {2,6} </center><br> Se hacen las correspondientes transiciones: <br>
<center>($2a$)   &delta;(2, c) = {4} </center><br>
<center>($2b$)   &delta;(6, c) = {5} </center><br><br>

Con esto podemos darnos cuenta que al hacer la función de transición extendida para nuestra palabra $ac$, los estados obtenidos fueron {4,5}, de los cuales ninguno corresponde a $3$, que es el único estado de aceptación de este autómata, lo que nos permite concluir que la afirmación es <b>FALSA</b>.

      Algunas de las ventajas que puede tener usar condicinales frente a máquinas de estado son que, con base en ciertos estados, se pueden llevar a cabo la toma de decisiones, esto con el fin de saber que acción conviene más en momentos determinados, además de dar un rango de opciones y acciones mucho más amplio.<br><br>
      
  <li>¿Cómo se utilizan los autómatas en el diseño de personajes para videojuegos?</li><br>
      Los autómatas, dada su definición, se pueden implementar dentro de los videojuegos como el campo de acción de un personaje dentro del juego, es decir, tiene un estado inicial, ya que todo juego tiene un punto de partida, por lo tanto lo debe tener el personaje a manejar; tiene un conjunto de estados, los cuales corresponderán a las diferentes formas en las que se podría mover o reaccionar un personaje; tiene un conjunto de estados de aceptación, que corresponderán a los movimientos exitosos; tiene un alfabeto determinado, que corresponderá a las acciones que puede realizar en algunas ocaciones; y por supuesto tendrá una función de transición, la que definirá el campo de acción que puede tener un personaje despues de ejecutar un movimiento (elemento del alfabeto). <br><br> También los autómatas se podrían implementar en el entorno o variables dentro del juego que puedan aceptar la historia, o modo de juego que se esté llevando a cabo.<br><br>
